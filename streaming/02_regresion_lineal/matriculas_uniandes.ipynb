```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../../data/admitidos_uniandes.csv")
```

```python
progs = list(set(data["PROGRAMA"]))
```

```python
data
```

```python
for p in progs:
    ii = data["PROGRAMA"]==p
    plt.plot(data["PERIODO"][ii]//100, data["MATRICULADOS"][ii], label=p)
plt.legend()
```
