# ¿Qué es una red neuronal?

## Unidad Neuronal

Las redes neuronales artificiales son el núcleo del DL. La primera red neuronal fue inspirada en el mecanismo de funcionamiento de las neuronas biológicas. La unidad fundamental de una red neuronal es la neurona. Tanto la red neuronal biológica como la red neuronal artificial están compuestas por esta unidad fundamental. Esta comparación suele ser un poco superficial y una mejor forma de entender una neurona es como si se tratase de una función matemática, que para una entrada dada se obtiene una salida. ## El Perceptron

El **perceptron** está compuesto por diferentes elementos en dos zonas de transformación. *Input* son los datos recibidos por el perceptron. Cada dato es ponderado por un peso (*weight*) $w_{n}$. Los productos entre las entradas $x$ y los pesos $w$ es sumado lineamente junto a un valor constante conocido como *bias*, así $z = \Sigma x_n w_{n} + b$ (Transformación lineal). El resultado pasa por una función no lineal, conocida como función de activación que se activará o no dependiendo del valor de la suma obteniendo una salida *output* (Transformación no-lineal). Esta activación nos permite saber si la salida coincide o no con lo esperado y para ello calculamos un error, definido como la diferencia entre lo esperado y lo obtenido, $\Delta = y - \bar{y}$. Este error nos sirve para actualizar los pesos. Cada vez que se obtiene una salida del perceptron se dice que ha transcurrido una época, después de cada época los pesos se actualizan así $w_n = w_n + \eta*\Delta*x_n$, donde $\eta$ es un parametro conocido como la tasa (rata) de aprendizaje (*learning rate*).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('percep_linear.csv')
data.head()
```

```python
X1 = data['X1']
X2 = data['X2']
Y = data['Y']

# Usamos máscaras para definir las clases
class_0 = Y == 0 
class_1 = Y == 1

#-- Graficamos los puntos con sus correspondientes clases
fig = plt.figure(figsize=(10,10))
plt.plot(X1[class_0],X2[class_0],'o', c='black', label='0', markersize=10)
plt.plot(X1[class_1],X2[class_1],'o', c='red', label='1', markersize=10)
plt.xlim(0,1)
plt.ylim(0,1)
plt.legend(fontsize=15)
plt.show()
```

```python
#-- Definimos la función de activación, en este caso usamos un Sigmoide
def act(z):
    return 1/(1 + np.exp(-z))
```

```python
#-- Esta es la función que queremos encontrar
def f(x):
    return -(b + w_1*x)/w_2
```

$$b + w_1\cdot x_1 + w_2 \cdot x_2 = 0$$

$$w_2 \cdot x_2 = - (b + w_1\cdot x_1 )$$

$$x_2 = - (b + w_1\cdot x_1 )/w_2$$

```python
# Inicializamos los pesos en 0
b = 0
w_1 = 0
w_2 = 0

# Inicializamos el número de épocas y la rata de aprendizaje
n_epochs = 15  # Número de épocas
lr = 0.5       # Tasa de aprendizaje

errors = []

# Entrenamos el Perceptron
for epoch in range(n_epochs):
    total_errors = 0
    
    for i,j,k in zip(X1,X2,Y):    
        # Función de Suma
        z = i*w_1 + j*w_2 + b

        # Función de Activación
        sig_z = act(z)

        # Evaluamos la salida (output)
        if sig_z >= 0.5:
            out = 1
        if sig_z < 0.5:
            out = 0

        # Calculamos el error
        error = k - out
        
        if error !=0:
            total_errors += 1

        # Actualizamos los pesos
        b = b + lr*error
        w_1 += lr*error*i
        w_2 += lr*error*j
    
    errors.append(total_errors)
    print('Epoch [{}/{}], bias: {}, w1: {}, w2: {}'.format(epoch+1,n_epochs,b,w_1,w_2))
    
    
    fig = plt.figure(figsize=(4,4))
    plt.plot(X1[class_0], X2[class_0],'o', c='black', label='0')
    plt.plot(X1[class_1], X2[class_1],'o', c='red', label='1')
    plt.plot(X1, f(X1),'-', c='green',)
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.legend()
    plt.show()
```

```python
fig = plt.figure(figsize=(8,8))
plt.plot(range(n_epochs), errors)
plt.xlabel('n_epochs')
plt.ylabel('Error')
plt.grid()
plt.show()
```
